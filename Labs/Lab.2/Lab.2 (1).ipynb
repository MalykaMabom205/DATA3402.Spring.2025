*Exercise 9:* Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

````python
import string

# Function to draw the board with row and column labels
def draw_board(board):
    """Draws the Tic Tac Toe board with labeled rows and columns."""
    n = len(board)
    # Print the column labels
    print("   " + "   ".join(string.ascii_uppercase[:n]))
    
    # Print each row with its corresponding label
    for i in range(n):
        row_str = f"{i+1} | " + " | ".join(["X" if cell == 1 else "O" if cell == 2 else " " for cell in board[i]]) + " |"
        print(row_str)
        print("   " + " ---" * n)  # Print the separator after each row

# Function to place a move on the board
def place_move(board, player, x, y):
    """Places a move on the Tic Tac Toe board."""
    n = len(board)

    if not (0 <= x < n and 0 <= y < n):
        return False  # Out of bounds

    if board[x][y] == 0:
        board[x][y] = player  # Place the move
        return True
    return False  # Position already occupied

# Function to place a move using labeled coordinates
def place_move_by_label(board, player, position):
    """Places a move on the board using labeled coordinates."""
    n = len(board)

    if len(position) < 2:
        return False  # Invalid format

    col_letter, row_number = position[0], position[1:]

    if col_letter.upper() not in string.ascii_uppercase[:n] or not row_number.isdigit():
        return False  # Invalid format

    row = int(row_number) - 1  # Convert row to zero-based index
    col = string.ascii_uppercase.index(col_letter.upper())  # Convert column letter to index

    return place_move(board, player, row, col)  # Call the place_move function

# Function to check the status of the game
def check_game_status(board):
    """Checks the game status: -1 for incomplete, 0 for draw, 1 for player 1 win, 2 for player 2 win."""
    n = len(board)

    # Check rows and columns for a winner
    for i in range(n):
        if board[i][0] == board[i][1] == board[i][2] != 0:
            return board[i][0]  # Winner in the row
        if board[0][i] == board[1][i] == board[2][i] != 0:
            return board[0][i]  # Winner in the column

    # Check diagonals for a winner
    if board[0][0] == board[1][1] == board[2][2] != 0:
        return board[0][0]  # Winner in the diagonal
    if board[0][2] == board[1][1] == board[2][0] != 0:
        return board[0][2]  # Winner in the diagonal

    # Check for a draw (no empty spaces)
    for row in board:
        if 0 in row:  # There are empty spaces, game is still incomplete
            return -1

    return 0  # Draw if no empty spaces and no winner

# Function to take a player's turn
def take_turn(board, player):
    """Prompts the player for a valid move and modifies the board."""
    while True:
        # Ask the player for a move (e.g., "A2", "B3")
        position = input(f"Player {player}, enter your move (e.g., 'A2'): ")

        # Call the place_move_by_label function to make the move
        if place_move_by_label(board, player, position):
            print(f"Player {player}'s move was successful!")
            break  # Exit the loop if the move is valid
        else:
            print("Invalid move. Please try again.")  # Ask for input again

# Main function to play the game
def play_game():
    """Plays the entire game of Tic Tac Toe."""
    board = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]  # Initialize an empty 3x3 board
    current_player = 1  # Player 1 starts the game

    while True:
        draw_board(board)  # Display the current board
        take_turn(board, current_player)  # Player takes a turn

        # Check if the game has ended
        status = check_game_status(board)
        if status == 1:
            draw_board(board)
            print("Player 1 wins!")
            break
        elif status == 2:
            draw_board(board)
            print("Player 2 wins!")
            break
        elif status == 0:
            draw_board(board)
            print("It's a draw!")
            break

        # Switch to the other player for the next turn
        current_player = 2 if current_player == 1 else 1

# Run the game
play_game()
````
